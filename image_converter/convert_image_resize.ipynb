```python
from PIL import Image
from google.colab import drive
from google.colab import files
import shutil
import os

# Mobile
# 1242 x 2208
# 1440 x 2560

# Tablet
# 2048 x 2732

google_drive_path = '/content/drive'

google_drive_mounted = os.path.exists('/content/drive')

# check google drive mounted
if google_drive_mounted == True:
    print('google drive mounted')
else:
    drive.mount(google_drive_path)
    print('google drive unmounted')

def resize_image(input_image_path, output_image_path, width, height):
    image = Image.open(input_image_path)
    origin_width, origin_height = image.size

    # BILINEAR, BICUBIC, LANCZOS, ANTIALIAS
    # BICUBIC, LANCZOS - High Quality
    resized_image = image.resize((width, height), Image.BICUBIC)
    resized_image.save(output_image_path)
    print(f"resize_image ... [{origin_width}x{origin_height}] >>> [{width}x{height}] converted!")

def move_to_drive_folder(output_file_path, output_file_name):
    main_folder = "/content/drive/MyDrive/donguran"
    if not os.path.exists(main_folder):
        os.mkdir(main_folder)
        print(f'move_to_drive_folder.. make main folder : {main_folder}')

    rpa_folder = "/content/drive/MyDrive/donguran/rpa"
    if not os.path.exists(main_folder):
        os.mkdir(rpa_folder)
        print(f'move_to_drive_folder.. make rpa folder : {rpa_folder}')

    target_folder = rpa_folder + "/resized"
    if not os.path.exists(target_folder):
        os.mkdir(target_folder)
        print(f'move_to_drive_folder.. make target folder : {target_folder}')

    output_drive_path = os.path.join(target_folder, output_file_name)
    try:
        shutil.move(output_file_path, output_drive_path)
        print(f"move_to_drive_folder.. move : [{output_file_path}] >>> [{output_drive_path}] success")
    except:
        print(f"move_to_drive_folder.. move : [{output_file_path}] >>> [{output_drive_path}] failed")


if __name__ == "__main__":
    # upload file
    uploaded = files.upload()

    while True:
        # user input width, height size.
        width = int(input("resize image width: "))
        height = int(input("resize image height: "))


        for input_file_name in uploaded.keys():
            input_file_path = os.path.join('/content', input_file_name)
            input_file_name = input_file_name.strip().replace(" ", "_")
            input_file_suffix = "." + input_file_name.rsplit('.', 1)[1]

            output_file_name = input_file_name.rsplit('.', 1)[0] + f"({width}x{height})" + input_file_suffix
            print(f"output_file_name:{output_file_name}")

            output_file_path = os.path.join('/content', output_file_name)
            resize_image(input_file_path, output_file_path, width, height)

            move_to_drive_folder(output_file_path, output_file_name)

        # add - suggest other size to images
        while True:
            isRepeat = input("if you want other size, Enter 'y' or 'n' or 'upload' : ").lower()

            if isRepeat == 'y':
                break
            elif isRepeat == 'n':
                break
            elif isRepeat == 'upload':
                uploaded = files.upload()
                break

        if isRepeat == 'n':
            break
```
